```python
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib widget
```

```python
input_file = '/Users/giadaaggio/Desktop/Thesis/TOTORO/FITS/F225W/MAT.004'
data = pd.read_csv(input_file, comment='#', delim_whitespace=True, header=None, \
                    names=['0', '1', '2', '3', '4', '5', 'res1', 'res2', 'res3', 'res4', '10', '11', '12', '13', '14'])
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data['res1'], data['res2'], s=1, c='black')
plt.scatter(data['res3'], data['res4'], s=1, c='black')
plt.axvline(0.0, color='red', linestyle='--', alpha=0.9, linewidth=0.5)
plt.axhline(0.0, color='red', linestyle='--', alpha=0.9, linewidth=0.5)
plt.xlim(-0.1, 0.1)
plt.ylim(-0.1, 0.1)
plt.xlabel(r'$\Delta x$ (px)')
plt.ylabel(r'$\Delta y$ (px)')
plt.title('Residuals')
plt.tight_layout()
plt.show()
```
