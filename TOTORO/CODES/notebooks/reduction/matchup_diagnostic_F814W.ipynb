```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
input_file = '/Users/giadaaggio/Desktop/Thesis/TOTORO/FITS/F814W/MATCHUP.XYMEEE'
data = pd.read_csv(input_file, comment='#', delim_whitespace=True, header=None, \
                    names=['x', 'y', 'mag', 'res_x', 'res_y', 'res_mag', 'qfit', '7', '8', '9', '10', '11', '12', '13'])

data_selected = data #[(data['mag']<-3)]
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(data_selected['x'], data_selected['y'], c=data_selected['mag'], s=0.1, cmap='magma_r')
cbar = plt.colorbar()
cbar.set_label('Magnitude')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Field F225W')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(data_selected['mag'], data_selected['res_mag'], s=0.5, alpha=0.5, color='black')
plt.xlim(-17.5, -5)
plt.ylim(-0.1, 1.)
plt.xlabel('Magnitude')
plt.ylabel('Residuals')
plt.title('Magnitude vs residuals')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.scatter(data_selected['x'], data_selected['res_x'], s=0.5, alpha=0.5, color='black')
plt.xlabel('X')
plt.ylabel('Residuals X')
plt.title('X residuals')

plt.subplot(122)
plt.scatter(data_selected['y'], data_selected['res_y'], s=0.5, alpha=0.5, color='black')
plt.xlabel('Y')
plt.ylabel('Residuals Y')
plt.title('Y residuals')

plt.tight_layout()
plt.show()
```
